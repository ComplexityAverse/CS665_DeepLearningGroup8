function to perform labels from csv file

```python
def assign_num(num):
    if (1 <= num <= 3):
        return 0
    if (4 <= num <= 8):
        return 1
    if (9 <= num <= 16):
        return 2
    if (17 <= num <= 25):
        return 3
    if (26 <= num <= 32):
        return 4
    if (33 <= num <= 42):
        return 5
    if (43 <= num <= 55):
        return 6
    if (56 <= num <= 66):
        return 7
    if (67 <= num <= 110):
        return 8
```
